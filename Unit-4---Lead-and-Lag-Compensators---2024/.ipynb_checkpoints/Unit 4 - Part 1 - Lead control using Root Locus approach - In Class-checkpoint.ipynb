# What is the form of the Lead and Lag Compensator?

Lead Compensator
$$
K(s) = K_{ld}\frac{s+b}{s+a}, \hspace{1cm} b<a
$$

Lag Compensator
$$
K(s) = K_{lg}\frac{s+b}{s+a}, \hspace{1cm} b>a
$$

# When do we use Lead and Lag Compensators?

When PID (or other current simple controller) is not quite good enough. 

Lead and Lag compensators are ususally added in-line with another simple controller.

![closedloop](closedloop.png "Closed Loop")

General philosophy: start simple, then gradually increase complexity if needed.

Lead controller:
- speed up system
- increase robustness
- increase phase margin


Lag controller:
- slow down system
- improve reference tracking
- improve disturbance rejection
- keep transient response the same, but reduce steady state error

An advantage of Lead and Lag Compensators is that they are easy to build in hardware (using op-amps, resistors, capacitors and inductors).

# Lead Compensators - Design using Root Locus 

This method highlights how to speed up the response of a closed-loop system using a Lead Compensator

# Preliminary Material

Consider the closed-loop transfer function $T(s)$:
$$
T(s) = \frac{P(s)K(s)}{1 + P(s)K(s)}.
$$
From this expression we see that the closed loop poles of the system satisfy the equation
$$
1 + P(s)K(s) = 0,
$$
or
$$
P(s)K(s) = -1.
$$
Some observations:
- $L(s) = P(s)K(s)$ is a complex number, so we can write it in polar form as $M_{L}(s) e^{j \theta_L(s)}$.
- $L(s)$ is a rational function, so the magnitude and phase can be calculated as:
$$
\begin{align*}
L(s) &= \frac{(s+z_1)\cdots(s+z_{n_z})}{(s+p_1)\cdots(s+p_{n_p})}\\
&= \frac{M_1(s)e^{j \theta_1(s)} \cdots M_{n_z}(s)e^{j \theta_{n_z}(s)}}{N_1(s) e^{j \phi_1(s) } \cdots N_{n_p}(s)e^{j \phi_{n_p}(s)}} \\
&= \frac{M_1(s) \cdots M_{n_z}(s)}{N_1(s) \cdots N_{n_p}(s)}e^{j (\theta_1(s) + \cdots \theta_{n_z}(s) - \phi_1(s) - \cdots - \phi_{n_p}(s) ) }
\end{align*},
$$
which means that:
$$
\begin{align*}
M_L(s) &= \frac{M_1(s) \cdots M_{n_z}(s)}{N_1(s) \cdots N_{n_p}(s)} \\
\theta_L(s) &= \theta_1(s) + \cdots \theta_{n_z}(s) - \phi_1(s) - \cdots - \phi_{n_p}(s) 
\end{align*}
$$
- $-1$ can be written in polar form as $1 e^{j (\pi \pm 2k \pi)}$, $k=0, 1, 2, \ldots$.

Let $-a_1$, $-a_2$, $\ldots$, $-a_{n_p}$ be the closed loop poles of the system. By equating the angles and magnitudes of both sides of the above equations we find that the closed loop poles satisfy the following conditions:
- Angle Condition:
    $$
    \theta_L(-a_i) = \pi \pm 2k \pi, \hspace{1cm} k=0, 1, 2, \ldots, i=1,\ldots,n_p
    $$
- Magnitude Condition:
    $$
    M_{L}(-a_i) = 1, \hspace{1cm} i = 1, \ldots, n_p
    $$


# Key Insight

All points on the Root Locus represent the closed-loop poles for different values of gain. Therefore, all points on the Root Locus satisfy the Angle and Magnitude Conditions.

# How to design a Lead Compensator Using the Root Locus Approach

Main idea: 
- Select closed-loop poles that lie to the left of the current dominant closed-loop poles
- Shape the Root Locus so that it satisfies the Angle and Magnitude Conditions at the desired closed-loop poles

# Illustrate the Idea By Example

Consider the plant $P(s) = \frac{s+2}{(s+3)(s+1+j)(s+1-j)}$. Suppose that we want to place the dominant closed loop poles at $-2 \pm j3$. The Root Locus for the $P(s)$ is shown below. Note that the desired closed-loop poles do not lie on the Root Locus (red crosses). 

```python
import numpy as np
import math
import control as ct 
import enel441_utilities as eu
import matplotlib.pyplot as plt

```

```python
P_poles = np.array([-3, -1-1j, -1+1j])
P_zeros = np.array([-2])
den = eu.roots_to_polynomial(P_poles)
num = eu.roots_to_polynomial(P_zeros)
P = ct.tf(num,den)

fig, ax = plt.subplots(1)
cl_poles, gain_vec = ct.root_locus(P)
ax.plot(-2,3,'rx')
ax.plot(-2,-3,'rx')
```



If we want $-2 \pm j3$ to lie on the Root Locus, then we must ensure that the angle of the Loop Transfer function is 180 at $s=-2 \pm j3$ and the magnitude is $1$. 

$$
\begin{align*}
P(s)K(s) \Big |_{s=-2 \pm j3 } = -1
\end{align*}
$$

Remember that the form of the Lead Compensator is $K(s) = \frac{s+b}{s+a}$. So we have:
$$
\begin{align*}
\frac{s+2}{(s+3)(s+1+j)(s+1-j)} \frac{s+b}{s+a} \Big |_{s=-2 \pm j3 } &= -1 \\
\frac{-2 \pm j3 +2}{(-2 \pm j3 +3)(-2 \pm j3 +1+j)(-2 \pm j3 +1-j)} \frac{-2 \pm j3 +b}{-2 \pm j3 +a} &= -1 \\
\end{align*}
$$

Now we just need to solve for $a$ and $b$ to satisfy this equation. Note that $a$ and $b$ usually are not unique. There are many solutions that are possible for this equation. 

First, we need to know what the angle is at $s=-2 \pm j3$ for the uncompensated system:
$$
\begin{align*}
P(-2+j3) &= \frac{-2+j3+2}{(-2+j3+3)(-2+j3+1+j)(-2+j3+1-j)} \\
&= \frac{3\angle{90}}{(\sqrt{10}\angle{71.56})(\sqrt{17}\angle{104})(\sqrt{5}\angle{116})} \\
&= \frac{3}{\sqrt{10}\sqrt{17}\sqrt{5}}\angle 90 - 71.56 - 104 - 116 \\
&= 0.1029 \angle -201.56
\end{align*}
$$
So if we want an angle of -180, we need to add $201.56-180 = 21.56$ degrees to the Loop Transfer function. 

```python
# Code use for the above calculations:

# pole at s=-3:
print('Pole at s=-3:')
print('phase=', math.atan2(3,1)*180/np.pi)
print('mag =', np.sqrt(3**2+1))

# pole at s=-1+j:
print('\nPole at s=-1+j:')
print('phase=', math.atan2(2,-1)*180/np.pi)
print('mag =', np.sqrt(2**2+1))

# pole at s=-1-j:
print('\nPole at s=-1-j:')
print('phase=', math.atan2(4,-1)*180/np.pi)
print('mag =', np.sqrt(4**2+1))

# zero at s=-2:
print('\nZero at s=-2:')
print('phase=', math.atan2(3,0)*180/np.pi)
print('mag =', np.sqrt(3**2))
```

# Visual Interpretation

Consider a the numerator $(s+2)$, evaluated at $s=-a$. What geometrical operation are we doing?

- When we write $(-a+2)$, we are subtracting two points in Euclidean space. 
- Subtracting two points defines a line. In this case the line is from the desired closed loop pole to the open loop zero. 
- Converting to polar co-ordinates is equivalent to expressing the line as an angle and magnitude.

![splane](splanePannotated.png "S-Plane")

# Changing the Angle

We will add $21.5$ degrees of phase to the point $-2+j3$ by adding a real pole and a real zero. 

Let's add a zero at $-5$. This is the zero of the Lead Compensator.

We need to add a bit more than $21.5$ degrees since when we place the pole, we will lose a bit of phase.

Let's calculate the angle added by the zero at $-5$ to the current point $-2+j3$:
$$
\begin{align*}
(s+5)\Big |_{s=-2+j3} &= -2 + j3 + 5 \\
&= \sqrt{18}\angle 45
\end{align*}
$$

![splane](splanePannotated2.png "S-Plane")

```python
# zero at s=-5:
print('\nZero at s=-5:')
print('phase=', math.atan2(3,3)*180/np.pi)
print('mag =', np.sqrt(3**2+3**2))
```

When we include the zero at $-5$ in the open loop transfer function the total angle at the point $s=-2+j3$ is:
$$
\theta_L = 45 - 201.56 = -156.56 \text{ degrees}
$$
Recall that we want the phase of the loop transfer function to be -180 at the point $s=-2+j3$ to ensure that this point falls on the Root Locus of the compensated system.

So now we need to subtract $180-156.56=23.44$

We will do this using a pole. We just need to find the correct location on the real axis in order to remove exactly $23.44$ degrees of phase at the point $s=-2+j3$.

Graphically we have the following situation:

![splane](splanePannotated3.png "S-Plane")

We need to solve for $x$ in the figure to determine the pole location. We have the following equation:
$$
\tan(23.4) = \frac{3}{x}.
$$
From which we find that $x = 6.92$. Thus the pole location for the lead compensator is $-8.92$.

Finally, the Lead Compensator that we designed has the form:
$$
K(s) = K_{ld}\frac{s+5}{s+8.92}.
$$
Thus the loop transfer function is:
$$
\begin{align*}
L(s) &= K(s) P(s) \\
&= K_{ld}\frac{s+5}{s+8.92} \frac{s+2}{(s+3)(s+1+j)(s+1-j)}.
\end{align*}
$$

The Root Locus of the loop transfer function is show below. Note that now the desired closed-loop poles now lie on the Root Locus!

```python
L_poles = np.array([-8.92, -3, -1-1j, -1+1j])
L_zeros = np.array([-5, -2])
den = eu.roots_to_polynomial(L_poles)
num = eu.roots_to_polynomial(L_zeros)
L = ct.tf(num,den)
  
fig, ax = plt.subplots(1)
cl_poles, gain_vec = ct.root_locus(L)
ax.plot(-2,3,'rx')
ax.plot(-2,-3,'rx')
```

# Determining the gain $K_{ld}$

The last step is to determine the gain of the lead compensator. 

Recall the gain condition, that any point $a$ on the Root Locus must satisfy $|L(s)|_{s=a} = 1$.

We can use this formula to solve for $K_{ld}$:
$$
\begin{align*}
|L(-2+j3)| &= K_{ld} \frac{\sqrt{18}}{7.54} \frac{3}{\sqrt{10}\sqrt{17}\sqrt{5}} \\
1 &= K_{ld} 0.0579 \\
K_{ld} &= 17.27
\end{align*}
$$

This completes the design of the Lead Compensator!
