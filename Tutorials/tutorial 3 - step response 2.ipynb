# Question 1

To calculate the steady state value of a step response use the formula:
$$
\begin{align*}
y(\infty) = \lim_{s \to 0} s Y(s)
\end{align*}
$$

When applied to the step response in the question we have:
$$
\begin{align*}
y(\infty) &= \lim_{s \to 0} s \frac{b_0 s + b_1}{a_0 s^2 + a_1 s + a_2} \frac{1}{s} \\
&= \frac{b_1}{a_2}
\end{align*}
$$
